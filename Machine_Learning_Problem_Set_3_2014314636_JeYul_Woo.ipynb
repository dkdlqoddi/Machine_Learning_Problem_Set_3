```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

inputs = list()
inputs.append(np.loadtxt("./hw3_data/hw3_2_X_1.csv", delimiter=',', dtype=np.float32))
inputs.append(np.loadtxt("./hw3_data/hw3_2_X_2.csv", delimiter=',', dtype=np.float32))
inputs.append(np.loadtxt("./hw3_data/hw3_2_X_3.csv", delimiter=',', dtype=np.float32))
print(inputs[0].shape)
print(inputs[1].shape)
print(inputs[2].shape)

labels = list()
labels.append(np.loadtxt("./hw3_data/hw3_2_Y_1.csv", delimiter=',', dtype=np.float32))
labels.append(np.loadtxt("./hw3_data/hw3_2_Y_2.csv", delimiter=',', dtype=np.float32))
labels.append(np.loadtxt("./hw3_data/hw3_2_Y_3.csv", delimiter=',', dtype=np.float32))
print(labels[0].shape)
print(labels[1].shape)
print(labels[2].shape)

clf = svm.SVC(kernel='linear')
clf.fit(inputs[0], labels[0])
output = clf.predict(inputs[0])

plt.scatter(inputs[0][:,0], inputs[0][:,1], c=labels[0], s=labels[0].shape[0], cmap=plt.cm.Paired)
ax = plt.gca()

xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx = np.linspace(xlim[0], xlim[1], labels[0].shape[0])
yy = np.linspace(ylim[0], ylim[1], labels[0].shape[0])
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

ax.contour(XX, YY, Z, colors='k', levels=[-1,0,1], alpha=0.5, linestyles=['--', '-', '--'])

ax.scatter(clf.support_vectors_[:,0], clf.support_vectors_[:,1], s=labels[0].shape[0]*2, facecolors='r')

plt.show()
```
